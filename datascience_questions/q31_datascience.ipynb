$\huge ρ_{X,Y} = \frac{COV(X,Y)}{σ_X σ_Y} = \frac{E[(X-\mu_x)(Y-\mu_y)]}{σ_X σ_Y}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.normal(loc=10, scale=2, size=1_000)
Y = np.random.normal(loc=10, scale=10, size=1_000)
#Y = 2*X + np.random.normal(loc=0, scale=1, size=1_000)
```

```python
plt.scatter(X, Y)
```

```python
((X - np.mean(X))*(Y - np.mean(Y))).mean() / (np.std(X) * np.std(Y)), ((X - np.mean(X))*(Y - np.mean(Y))).mean(), (np.std(X) * np.std(Y))
```
