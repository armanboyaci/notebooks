```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bounds = [(0, 1), (-1, 2), (-2, 3), (-1,1), (-10, 10)]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(20,8))
fig.suptitle(f"$Y=X^2$", fontsize=40)

for i, bound in enumerate(bounds):
  low, high = bound
  X = np.random.uniform(low=low, high=high, size=10_000)
  Y = X**2
  corr = np.corrcoef(X, Y)[0, 1]
  cov = np.cov(X, Y)[0, 1]

  ax[0, i].scatter(x=X, y=Y, s=3)
  ax[0, i].set_title(f"Corr: {corr:.2f}, Cov: {cov:.2f}")

  Y = Y  + np.random.normal(0, 0.1, size=10_000)
  corr = np.corrcoef(X, Y)[0, 1]
  cov = np.cov(X, Y)[0, 1]

  ax[1, i].scatter(x=X, y=Y, s=3)
  ax[1, i].set_title(f"Corr: {corr:.2f}, Cov: {cov:.2f}")
```

```python
from sklearn.datasets import make_circles

X, _ = make_circles(random_state=42, n_samples=300)
Y = X[:, 1]
X = X[:, 0]

corr = np.corrcoef(X, Y)[0, 1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))

plt.scatter(X, Y)
plt.title(f"corr: {corr:.2f}")
```
