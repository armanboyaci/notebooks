**Question**

How can you decide how long to run an experiment? What are some problems with just using a fixed p-value threshold and how do you work around them?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
```

```python
mu    = [0, 0]
sigma = [0.40, 0.80]
N     = [10_000, 10_000] # population size
n     = [1000, 2000]     # sample size
```

```python
X = pd.Series(np.random.normal(loc=mu[0], scale=sigma[0], size=N[0]))
X.hist(bins=100)

Y = pd.Series(np.random.normal(loc=mu[1], scale=sigma[1], size=N[1]))
Y.hist(bins=100, color="red", alpha=0.5,  figsize=(10,5))
```

```python
fig, axs = plt.subplots(3, 3, sharex=True, figsize=(18,6))

sigma = [0.50, 1.00]

X = pd.Series(np.random.normal(loc=mu[0], scale=sigma[0], size=N[0]))
Y = pd.Series(np.random.normal(loc=mu[1], scale=sigma[1], size=N[1]))

n     = [1000, 2000]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[0,0])
axs[0][0].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)

n     = [100, 200]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[1,0])
axs[1][0].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)

n     = [10, 20]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[2,0])
axs[2][0].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)


sigma = [1.00, 2.00]

X = pd.Series(np.random.normal(loc=mu[0], scale=sigma[0], size=N[0]))
Y = pd.Series(np.random.normal(loc=mu[1], scale=sigma[1], size=N[1]))

n     = [1000, 2000]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[0,1])
axs[0][1].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)

n     = [100, 200]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[1,1])
axs[1][1].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)

n     = [10, 20]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[2,1])
axs[2][1].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)

sigma = [1.00, 1.00]

X = pd.Series(np.random.normal(loc=mu[0], scale=sigma[0], size=N[0]))
Y = pd.Series(np.random.normal(loc=mu[1], scale=sigma[1], size=N[1]))

n     = [1000, 2000]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[0,2])
axs[0][2].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)

n     = [100, 200]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[1,2])
axs[1][2].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)

n     = [10, 20]
mean_differences = [np.random.choice(X.values, n[0]).mean() - np.random.choice(Y.values, n[1]).mean() for _ in range(1_000)]
lb, ub = np.round(pd.Series(mean_differences).quantile(0.05), 2), np.round(pd.Series(mean_differences).quantile(0.95), 2)
pd.Series(mean_differences).hist(bins=100, ax=axs[2,2])
axs[2][2].set_title(f"n: {n}, $\sigma$: {sigma}, b=[{lb}, {ub}]", fontsize = 10)


plt.tight_layout()
```
